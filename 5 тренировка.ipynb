```python
n = int(input())

mas_mountain = [0] * n
for one in range(n):
    mas_mountain[one] = list(map(int, input().split()))
    
k = int(input())



mas_forward = list()
mas_back = list()


mas_forward.append(0)
mas_back.append(0)


for one in range(1, n):
    if mas_mountain[one][1] > mas_mountain[one -1][1]:
        mas_forward.append(mas_mountain[one][1] - mas_mountain[one -1][1] + mas_forward[one - 1])
    else:
        mas_forward.append(mas_forward[one - 1])

mas_mountain1 = mas_mountain[::-1] 
for one in range(1, n):
    if mas_mountain1[one][1] > mas_mountain1[one -1][1]:
        mas_back.append(mas_mountain1[one][1] - mas_mountain1[one -1][1] + mas_back[one - 1])
    else:
        mas_back.append(mas_back[one - 1])
mas_back = mas_back[::-1]

answers = list()

questions =list()

for on in range(k):
    first, second = list(map(int, input().split()))
    
    if first < second:
        answers.append(mas_forward[second -1] -mas_forward[first -1])
    elif first > second :
        answers.append(mas_back[second - 1] -mas_back[first -1])
    else:
        answers.append(0)
            
for another in answers:
    print(another)
```
